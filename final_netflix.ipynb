##6. from collections import Counter

The Counter class from the collections module counts the frequency of elements in an iterable (like lists or strings) and returns a dictionary-like object with elements as keys and their counts as values. It’s useful for tasks like finding the most common items or quickly tallying occurrences.

### Mapping of month numbers to month names

```python

month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Apply the mapping to the 'month' column
df['month'] = df['month'].map(month_mapping)

# Check the changes
print(df[['month']].head())
```

### Normalize duration_type values

```python

df['duration_type'] = df['duration_type'].replace({'Season': 'Seasons'})

# Double-check unique values to confirm
print(df['duration_type'].unique())
```

### Identify rare ratings — let's set a threshold for 'rare' ratings

```python
from sklearn.preprocessing import LabelEncoder

# Identify rare ratings — let's set a threshold for 'rare' ratings
rating_counts = df['rating'].value_counts()
rare_ratings = rating_counts[rating_counts < 10].index  # Adjust threshold as needed

# Replace rare ratings with 'Other'
df['rating'] = df['rating'].replace(rare_ratings, 'Other')

# Initialize LabelEncoder
le = LabelEncoder()

# Columns to encode
cols_to_encode = ["rating", "month"]

# Apply LabelEncoder to the specified columns
for col in cols_to_encode:
    df[col] = le.fit_transform(df[col])

# Check if outliers are handled and encoding worked
print(df[['rating', 'month']].head())
```

### IDENTIFY THE OUTLIERS

```python
sns.boxplot(df['rating'])
plt.show()
```

### HANDLE THE OUTLIERS

```python


# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Clip values outside the bounds
df['rating'] = df['rating'].clip(lower=lower_bound, upper=upper_bound)
```

```python
sns.boxplot(df['rating'],color="purple")
plt.show()
```

```python
Q1 = df['duration_num'].quantile(0.25)
Q3 = df['duration_num'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Outliers
outliers = df[(df['duration_num'] < lower_bound) | (df['duration_num'] > upper_bound)]
print(f"Number of outliers: {outliers.shape[0]}")
```

```python
df_no_outliers = df[(df['duration_num'] >= lower_bound) & (df['duration_num'] <= upper_bound)]
print(f"Data without outliers: {df_no_outliers.shape[0]} rows")


```

```python

df['duration_num'] = np.where(df['duration_num'] > upper_bound, upper_bound,
                             np.where(df['duration_num'] < lower_bound, lower_bound, df['duration_num']))
```

## NUMBER OF TITLES BY COUNTRY

```python
country_data = df['country'].value_counts().reset_index()
country_data.columns = ['country', 'count']

# Plot a choropleth map to visualize the number of titles by country
fig = px.choropleth(
    country_data,
    locations='country',
    locationmode='country names',
    color='count',
    hover_name='country',
    color_continuous_scale='Viridis',
    title='Number of Titles by Country'
)

# Customize layout with larger dimensions
fig.update_layout(
    width=1200,
    height=800,
    geo=dict(showframe=False, showcoastlines=True, projection_type='equirectangular'),
    coloraxis_colorbar=dict(title="Number of Titles")
)

fig.show()
```
